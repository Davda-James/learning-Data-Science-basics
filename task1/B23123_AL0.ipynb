3.1. Print the sum of left diagonal and right diagonal of a square matrix. 

```python
import numpy as np
mat= np.array([[1,3,4,5,2],[1,5,2,4,3],[5,2,3,4,1,],[1,4,2,6,9],[4,5,2,1,7]])

def left_diagonal_sum(arr): 
    #write your logic here
    n=arr.shape[0]
    ans=0
    for i in range(n):
        ans+=arr[i][i]
    return ans


def right_diagonal_sum(arr):
    #write your logic here
    n=len(arr)
    ans=0
    for i in range(n):
       ans+=arr[i][4-i]
    return ans


print(f'Left Diagonal Sum of {mat=} is {left_diagonal_sum(mat)}')
print(f'Right Diagonal Sum of {mat=} is {right_diagonal_sum(mat)}')
```

I sliced the array and find the left diagonal and right diagonal sum of array as mentioned in function using python property ':'

Question 4:  array operations in np.array and np math functions. Plot the function: $$ f(x) = {sin^7(x) + cos^5(x) \over e^x}$$ in the  domain: $$x \in [0,4]$$ Do not use for loops for the same. 

```python
import matplotlib.pyplot as plt
import numpy as np
def f(x :float) ->float:
    return (np.sin(x)**7 +np.cos(x)**5)/np.exp(x) 
     
steps =0.01
x= np.arange(0,4,steps)
vectorize_fun=np.vectorize(f)
y=vectorize_fun(x)
plt.xlabel("X values")
plt.ylabel("f(x)")
plt.title(label="Plot f(x)=(sin^7(x) + cos^5(x))/e^x , x :[0,4]")
plt.plot(x,y)  
plt.show ()
```

Used vectorize function so that there is no need of using for loop and we can directly get the arr containing function value for corresponding x value and then we can plot it easily

Question 5. Create a subplot with two plots side by side. Plot a sine wave in the first subplot. Plot a cosine wave in the
second subplot. Add labels, titles, and a legend to the plots. 

```python
import matplotlib.pyplot as plt
import numpy as np
# for sine function
def sinf(x):
    return np.sin(x)
x=np.arange(-300,300,1)
x=list(map(lambda x:x*np.pi/50,x))
vectorize_sin=np.vectorize(sinf)
y=vectorize_sin(x)
fig,ax=plt.subplots(2,1)
ax[0].set_xlabel("X values-------->")
ax[0].set_ylabel("Sin(x)---------->")
ax[0].set_title("f(x)=sin(x)")
fig.suptitle("sine & cosine plot")
plt.subplot(1,1,1)
plt.grid(True)
plt.plot(x,y,color="green")
# for cos function
def cosf(x):
    return np.cos(x)
vectorize_cos=np.vectorize(cosf)
y2=vectorize_cos(x)
ax[1].set_xlabel("X values------->")
ax[1].set_ylabel("Cos(x)-------->")
ax[1].set_title("f(x)=cos(x)")
plt.subplot(2,1,2)
plt.plot(x,y2, color="red")
plt.grid(True)
plt.tight_layout()
plt.show()  
```

I wrote this code for question 5 of plotting sine and cosine graph side by side
I have used subplot in it to draw the two graphs in single page side by side 
Also used tight_layout, so that it fits into page without any extension or cutting of some part of graph
x,y label has been set also title is given
